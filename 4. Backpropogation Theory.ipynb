- computational graph


- Steps
    - forward pass (compute loss)
    - compute local gradients
    - backward pass (compute dLoss/dWeigths using the chain rule)

## Linear regression



- now minimize the loss
- dloss/dx

```python
import torch

x = torch.tensor(1.0)
y = torch.tensor(2.)
w = torch.tensor(1., requires_grad=True)

# forward pass
y_hat = w*x
loss = (y_hat - y)**2

print(loss)

# backward pass
loss.backward()
print(f'gradint : {w.grad}')

# next step
# upgrade the weight
# continue the same
```

```python
x = torch.tensor(1.0)
y = torch.tensor(2.)
w = torch.tensor(1., requires_grad=True)

for epoch in range(5):
    # forward pass
    print(f'epoch: {epoch}')
    y_hat = w*x
    loss = (y_hat - y)**2
    
    print(loss)
    
    # backward pass
    loss.backward()
    print(f'gradint : {w.grad}')
    
    with torch.no_grad():
        w = w-w.grad
        
    print(f'w: {w}')    
    w.grad.zero_()
```
