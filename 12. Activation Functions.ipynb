# Activation function

## Why activation function
- y = w.X + b
- if we are not using an activation function then the output will be always a linear trasfrom of the input. This is not suited for compex tasks
- with activation
    - y = softmax(w.X+b)

## popular activation functions
- step function
- sigmoid
- tanh
- ReLU
- Leeky Relu
- Softmax
